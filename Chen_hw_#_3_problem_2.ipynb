The object siding down the incline is influenced under the gavitational force and resistance force. The equation of motion of the object is 

$\ddot{x}=-g\sin\theta+k\left(\frac{dx}{dt}\right)^{2}$

The dimsionless variables is defined by$ X=kx$ and $T=\sqrt{gk}t$. The new equation will be

$\ddot{X}=-\sin\theta+\left(\frac{dX}{dT}\right)^{2}$

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def f_int(y, x, theta):
    '''Defines the function'''
    #g=9.8
   # k=3
    return [(np.sin(theta))-y[0]**2, y[0]]

def analytic(t,theta):
    #g=9.8
    return np.log(np.cosh(np.sqrt(np.sin(theta))*t))

y_0 = np.array([0,0])
x=np.linspace(0.0, 10.0,1000000)
y_sol = odeint(f_int, y_0, x, args=(np.pi/6,))[:,1]
plt.figure(figsize=(9,7),dpi=98)
plt.plot(x, y_sol,linewidth=1)
plt.plot(x,analytic(x,np.pi/6),dashes=[5, 5, 5, 5],linewidth=2)
plt.grid(True)
plt.xlabel('T ($\sqrt{gk}t$)',fontsize=10)
plt.ylabel('X ($kx$)',fontsize=10)
plt.legend()
plt.show()
```
