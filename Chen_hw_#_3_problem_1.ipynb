```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

The Lagrangian of pendulam motion is  

$L=T-V=\frac{1}{2}ml^{2}\left(\dot{\theta}\right)^{2}-mgl\left[1-\cos\left(\theta\right)\right]$

Therefore, we can get the Eular Lagrangian equation

$L	=T-V=ml^{2}\left(\ddot{\theta}\right)+mgl\sin\left(\theta\right)
\longrightarrow	\ddot{\theta}=-\frac{g}{l}\sin\left(\theta\right)$

```python
def f_int(y, x,l):
    '''Defines the function'''
    g=9.8
    return [-g*np.sin(y[1])/l, y[0]]

def y_approx(x,l,y0):
    g=9.8
    return y0*np.cos(x*np.sqrt(g/l))
```

```python
x = np.linspace(0.0, 20.0, 1000000)
y_0 = np.array([0.0, np.pi/18])
y_sol = odeint(f_int, y_0, x,args=(10,))[:, 1]
plt.figure(figsize=(9,7),dpi=98)
plt.plot(x, y_sol,label='The numerical solution',linewidth=1)
plt.plot(x, y_approx(x,10,np.pi/18),label='The approximation',linewidth=1)
plt.grid(True)
plt.legend()
plt.show()
```
